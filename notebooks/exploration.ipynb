# Atari RL Environment Exploration

This notebook demonstrates how to interact with the Atari environment and visualize gameplay.

```python
import sys
sys.path.insert(0, '..')

import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output

from environments.atari_wrapper import make_atari_env
from agents.random_agent import RandomAgent
```

## Create Environment

```python
env = make_atari_env("Pong", frame_stack=4)
print(f"Observation space: {env.observation_space}")
print(f"Action space: {env.action_space}")
```

## Visualize Initial State

```python
obs, info = env.reset()

# Show stacked frames
fig, axes = plt.subplots(1, 4, figsize=(16, 4))
for i in range(4):
    axes[i].imshow(obs[i], cmap='gray')
    axes[i].set_title(f'Frame {i+1}')
    axes[i].axis('off')
plt.tight_layout()
plt.show()
```

## Run Random Agent

```python
agent = RandomAgent(env.action_space, env.observation_space)
obs, info = env.reset()

episode_reward = 0
steps = 0

terminated = False
truncated = False

while not (terminated or truncated) and steps < 1000:
    action = agent.select_action(obs)
    obs, reward, terminated, truncated, info = env.step(action)
    episode_reward += reward
    steps += 1
    
    if steps % 50 == 0:
        clear_output(wait=True)
        plt.figure(figsize=(6, 6))
        plt.imshow(obs[-1], cmap='gray')
        plt.title(f'Step {steps} | Reward: {episode_reward}')
        plt.axis('off')
        plt.show()

print(f"Episode finished after {steps} steps with reward {episode_reward}")
```
